```python
# 덧셈 연산

rgbColor = ['red', 'blue', 'green']
newColor = ['orange', 'yellow']

totalColor = rgbColor + newColor
print(totalColor)
```

```python
# 곱셈 연산

rgbColor = ['red', 'blue', 'green']
print(rgbColor * 2)

numbers = [0]
print(12 * numbers)
```

```python
# in 연산

rgbColor = ['red', 'blue', 'green']

print(f"green in rgbColor : {'green' in rgbColor}")
print(f"black in rgbColor : {'black' in rgbColor}")
```

```python
# 리스트 방문: 인덱스 값 사용

temps = [28, 31, 33, 35, 27, 26, 25]

for i in range(len(temps)) :
  print(temps[i], end = ' ')
```

```python
# 리스트 방문: for - each loop 사용

temps = [28, 31, 33, 35, 27, 26, 25]

for i in temps :
  print(i, end = ' ')
```

```python
# 리스트 값 확인

temps = [28, 31, 33, 35, 27, 26, 25]

if 33 in temps:
  print('temps에 33이 있다.')

if 39 not in temps:
  print('temps에 39가 없다.')
```

```python
# append()

colors = []
colors.append('black')
colors.append('white')
colors.append('yellow')
print(colors)
```

```python
# insert()

colors = ['red', 'green', 'blue']
colors.insert(1, 'white')
print(colors)
```

```python
# 리스트 탐색: index()

colors = ['red', 'green', 'blue']
if 'blue' in colors:
  print(colors.index('blue'))

  # 탐색하기 전 in 연산자를 이용하여 항목이 있는지 확인하는 것이 안전
```

```python
# 리스트 항목 삭제: pop()

colors = ['red', 'green', 'blue']
colors.pop(1)
print(colors)
```

```python
# 리스트 항목 삭제: remove()

colors = ['red', 'green', 'blue']
colors.remove('blue')
print(colors)
```

```python
# 최대값, 최소값, 합계

temps = [28, 31, 33, 35, 27, 26, 25]
print(f"합계 = {sum(temps)}")
print(f"최대값 = {max(temps)}")
print(f"최소값 = {min(temps)}")
```

```python
# 리스트 오름차순 정렬: sort()

temps = [28, 31, 33, 35, 27, 26, 25]
temps.sort()
print(temps)
```

```python
# 리스트 내림차순 정렬

temps = [28, 31, 33, 35, 27, 26, 25]
temps.sort(reverse=True)
print(temps)
```

```python
# random

import random
temps = [28, 31, 33, 35, 27, 26, 25]
print(f'랜덤하게 선택한 항목 = {random.choice(temps)}')

colors = ['red', 'green', 'blue']
print(f'랜덤하게 선택한 항목 = {random.choice(colors)}')
```

```python
# Ex 2. 최대값과 최소값 제외하기

scores = [10.0, 9.0, 8.3, 7.5, 5.0, 9.2, 6.4]
print(f'최대값과 최소값 제거 전: {scores}')

scores.remove(max(scores))
scores.remove(min(scores))

print(f'최대값과 최소값 제거 후: {scores}')
```

```python
# 리스트 비교

aList = [10, 20, 30, 40, 50]
bList = [10, 20, 30, 40, 50]
cList = [10, 20, 30, 40, 55]

print(f'aList == bList : {aList == bList}')
print(f'aList == cList : {aList == cList}')
```

```python
# 리스트 복사

x = [1, 2, 3]
y = x
y[0] = 10
print(x)
print(y)
```

```python
# 리스트 복사: copy()

x = [1, 2, 3]
y = x.copy()
y[0] = 10
print(x)
print(y)
```
